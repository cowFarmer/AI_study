이 꽃은 어떻게 표현(Representation)될까?   

1. 분자 형태; 가공을 거치지 않은 데이터 그대로
2. 이미지 형태; 픽셀을 RGB 형태로 값을 조절해서 시각적으로 표현
3. 표 형태; 데이터가 가진 특징(Feature)을 추출해서 각 열에 나타내고 값을 기입해서 표 형식으로 나타내기
4. 카테고리; 직접 사람이 정한 카테고리 안에서 어떤 카테고리인지 정하기 ex) ['차', '꽃', '고양이'] 에서 '꽃'에 해당함

1,2번 표현은 Raw, Sensory한 표현이라 말하고 3, 4번 표현은 Abstract, Internal한 표현이라 한다.

딥러닝은 사람의 신경망 구조를 모티브 삼아서 만들었다. 이는 데이터에서 내재적 표현(internal representations)을 찾아 모델을 학습시킬때 사용한다. 그래서 모델을 학습시킬만큼 데이터가 많이 필요하다.

# 딥러닝의 역사
딥러닝의 철학을 담은 역사를 알아보자   
> 행동주의: 학습된 행동은 자극의 반복으로 만들어진다.   

행동주의는 모든 지능은 반복된 자극으로 만들어진다고 주장했다.   
[스키너의 상자](http://www.edupolnews.com/news/articleView.html?idxno=11676)를 통해 보상을 주면 행동을 하는 강화학습(Reinforcement Learning)의 근원이 생겼다. 이러한 행동을 반복적으로 하면서 나중에는 보상을 얻기 위해 어떠한 행동을 하게 된다.~~파블로프의 개가 생각난다.~~   

> 인지주의: 인간은 자극을 받으면 행동하는게 아닌 뇌 안에서 정보처리 기능을 하는 의식이 있다.

행동주의와 다르게 학습, 기억 등의 정보처리 과정을 통해 의식하고 행동한다를 주장했다.   

> 연결주의: 정보처리하는 뇌에는 수 많은 뉴런의 연결고리가 있다.

인지주의에 있는 정보처리 과정에서 어떻게 설명하는지에 따라 다른 주의가 많이 있는데, 연결주의가 딥러닝의 근원이 되어 중요하다. 연결주의(Connectionism)은 수많은 신호들이 연결된 일련의 과정에 따라 정보가 처리된다고 주장한다.   

딥러닝이 연결주의의 뉴런이 얽혀있는 모습을 본뜬 인공 신경망을 모델로 가진다. 자극을 받아 내부에서 정보처리 과정을 거친 후 반응을 하는 모습이고, 정보처리를 하는 신경망 내부에서는 위에 있는 VGG16 모델 사진처럼 데이터를 여러 형태로 바꾸면서 표현하다가 우리가 의도한 형태의 데이터를 출력한다.

# 신경망의 본질
자극 -> 정보처리 -> 행동   

---

$x$input -> $f$ -> $y$output   
함수와 형태가 동일하다 = 신경망은 복잡한 알고리즘을 가진 함수이다.   
함수는 $y=f(x)$로 표현이 가능한데 $x$라는 입력값이 $f$에 들어가 $y$라는 값을 출력한다고 말할 수 있다.   

__함수의 역할__   
1. x와 y의 관계를 나타낼 수 있는 도구(Relation)   
$y=2x$의 경우 y는 x의 함수이다 라고 함의를 갖고 있다.y는 x의 변화에 종속적이고 x가 변하는 정도에 따라 y가 얼마나 변하는지는 x와 y의 함수의 형태로 결정된다.   
2. x를 변환해주는 도구(Transformation)   
변환의 관점에서 볼 때 선형 변환이 중요하다.
[3blue1brown youtube](https://www.youtube.com/watch?v=kYB8IZa5AuE)
여기 영상에서 보면 함수 대신 변환이라 한다. 그리고 변환 역할을 하는 함수가 변환 전 vector를 받아 변환 후 vector를 출력한다고 말하고 있다.   
변환은 x와 y축으로 이루어진 2차원 평면에서 축을 옮기거나 휘거나 꼬이게 할 수도 있다고 한

```python
# 변환의 관점
import numpy as np

def linear_transformation(X, A):
    transformed_vector = np.matmul(X, A)
    return transformed_vector

X = np.array([-1, 2])

A = np.array([[1, -2],
              [3, 0]])

# 1, 3 / *-1 *2 = 5
# -2, 0 / *-1 *2 = 2  -> [5, 2]

linear_transformation(X, A)
```

3. x의 공간에서 y의 공간으로 매핑해주는 도구(Mapping)   
위에서 나온 변환도 매핑이라고 볼 수도 있지만 이번에는 입력과 출력이 스칼라, 벡터에 따라 구분하는 매핑을 보려고 한다.   
스칼라: 크기만 갖고 있음, 벡터: 크기와 방향 두가지 갖고있고 좌표 상에서 화살표로 표현함   
세 가지 방법으로 입, 출력을 매핑할 수 있다.
- One to One   
x와 y 모두 스칼라인 경우이다. 예) 키 값이 있을 때 몸무게를 예측하는 함수   
키 -> f -> 몸무게 이를 선형 방정식으로 표현하면 $f(x)=wx+b=\hat{y}$   
여기서 $\hat{y}$는 예측값을 의미한다.   

- Many to One   
여러 가지 정보를 입력받아 하나의 값을 출력하는 경우이다.   
$x = [x_1, x_2 ... x_N]$ -> $f$ -> $y$   
위의 몸무게를 예측하는 함수에서 키 + 체지방, 골격근 등의 데이터를 입력해주면 키를 예측하기 더 용이해진다. 이는 벡터를 입력받아 하나의 스칼라를 출력하는 형태이다. 그리고 이러한 문제를 회귀라고 한다.(One to One도 포함)   

- Many to Many   
$x = [x_1, x_2 ... x_N]$ -> $f$ -> $y = [y_1, y_2 ... y_N]$   
예를들어 요즘 핫한 MBTI를 맞추는 모델은 문제를 통해 여러 성격의 특징을 모으고($x$) 함수를 통과한 후 성격유형과 퍼센트($y$)를 알려주는 형태. 이러한 문제가 분류 문제이다.

### 함수와 모델의 차이점
함수와 모델의 차이점은 무엇일까? 함수는 알고리즘이 고정된 형태이고 모델은 데이터를 받아 학습을 통해 원하는 값 혹은 형태를 출력하는 함수를 모델이라고 칭한다.   
또한 완벽한 함수를 찾게끔 모델을 학습시키는게 아니라 근삿값을 찾게끔 시도하는 것에 가깝다.   
어떻게 하면 잘 찾을 수 있을까?
1. 데이터를 보고 어떤 모델의 함수를 활용할지 정하기   
다양한 모델(함수)가 있는데 어떤 함수를 채택하냐에 따라 모델의 최종 성능에 중요한 영향을 미친다. 이미지나 NLP같은 비정형 데이터의 경우 선형함수가 아닌 신경망을 사용한다. 이의 특징은 여러 계층이 쌓인 형태인데 이런 구조가 데이터의 표현을 단계별로 잘 추출할 수 있기 때문이다.   
이렇게 데이터를 보고 함수 공간을 정하는 것을 Inductive Bias, Prior를 가정한다고 한다. 예를들어 함수 공간을 $f(x)=wx+b$같은 일차 함수 형태로 정해둔다면 데이터를 더 잘 표현할 수 있는 $f(x)=x^2+5x+2$같은 이차함수는 못찾을 수 있다. 그래서 예측을 잘 하려면 알맞은 함수 공간을 설정하는게 중요하다.   
예를들어 CNN, RNN, GAN, 의사결정나무 등 많이 있다.
2. 함수 형태를 고른 후 최적의 함수를 찾아가기   
이 단계가 모델 학습의 단계다 모델 학습시키는 방법도 모델의 형태에 따라 다양한데 예를들어 경사하강법(Gradient Descent), 역전파, SoftMax 등이 있다.
