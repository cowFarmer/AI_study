# 폐렴 진단기 만들기

딥러닝 기술을 의료 산업에 사용하면 숙련자 수준 이상의 정확도를 갖고있다. 영상분석 인력의 개인적인 편차, 주관적인 판단, 피로에 의한 오진 등 부정확성을 극복할 수 있는 대안으로 인정받고 있다. 단 의료 영상의 독특한 특징이 있다.
- 개인 정보 보호 등의 이슈로 데이터를 구하는 것이 쉽지 않다.
- 라벨링 작업이 전문적 지식을 요하므로 데이터셋 구축 비용이 비싸다.
- 희귀 질병을 다루는 경우 데이터 입수가 힘들다.
- 음,양성 데이터간 불균형이 심하다. 
- 이미지만으로 판단하기 어려워 다른 데이터와 결합해서 해석해야할 수 있다. 

그래서 의료 영상 처리를 위해 딥러닝 영상 처리 기술과 의료 도메인, 의료 영상에 대한 명확한 지식과 이해가 필요하다.

[kaggle Chest X-Ray Images(pneumonia)](https://www.kaggle.com/paultimothymooney/chest-xray-pneumonia) 데이터셋   
![](https://d3s0tskafalll9.cloudfront.net/media/original_images/12_e0QcAmT.png)   
폐렴은 말 그대로 폐에 염증이 생긴 상태이다. x-ray 이미지로 봤을 때(정상, 세균성 폐렴, 바이러스성 폐렴) 정상보다 염증이 있을때 흰색이 많이 있는걸 볼 수 있다. 이런 패턴을 딥러닝을 통해 학습시켜 진단을 내리는 모델을 만들어보자

## 1. 데이터 가져오기

```python
import os, re
import random, math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings(action='ignore')
```

```python
# AUTOTUNE 설정 변수
AUTOTUNE = tf.data.experimental.AUTOTUNE
# X-RAY 이미지 사이즈 변수
IMAGE_SIZE = [180, 180]

# 데이터 경로 변수
ROOT_PATH = os.path.join(os.getenv('HOME'), 'aiffel')
TRAIN_PATH = ROOT_PATH + '/chest_xray/data/train/*/*' # *은 모든 디렉토리와 파일을 의미한다.
VAL_PATH = ROOT_PATH + '/chest_xray/data/val/*/*'
TEST_PATH = ROOT_PATH + '/chest_xray/data/test/*/*'

# 하이퍼 파라미터 변수
BATCH_SIZE = 16
EPOCHS = 25

print(ROOT_PATH)
```

```python
train_filenames = tf.io.gfile.glob(TRAIN_PATH)
test_filenames = tf.io.gfile.glob(TEST_PATH)
val_filenames = tf.io.gfile.glob(VAL_PATH)

print(len(train_filenames))
print(len(test_filenames))
print(len(val_filenames))
```

train에 쓰일 val이 너무 없어서 데이터를 모은 후 8대2로 나누자.

```python
# filenames에 데이터 모으기
filenames = tf.io.gfile.glob(TRAIN_PATH)
filenames.extend(tf.io.gfile.glob(VAL_PATH))

# 모아진 filenames를 8:2로 나누기
train_size = math.floor(len(filenames)*0.8)
random.seed(8)
random.shuffle(filenames)
train_filenames = filenames[:train_size]
val_filenames = filenames[train_size:]

print(len(train_filenames))
print(len(val_filenames))
```

train 데이터 안에 있는 정상 이미지와 폐렴 이미지 수를 파악해보자.

```python
print(f'Normal image path\n{filenames[0]}')
print(f'Pneumonia image path\n{filenames[2000]}')
```

파일 경로에 NORMAL이 있으면 정상 PNEUMONIA가 있으면 폐렴 이미지이므로 이를 이용해서 수를 파악할 수 있는걸 확인했다.

```python
COUNT_NORMAL = len([filename for filename in train_filenames if "NORMAL" in filename])
print(f"Normal images count in training set: {COUNT_NORMAL}")


COUNT_PNEUMONIA = len([filename for filename in train_filenames if "PNEUMONIA" in filename])
print(f"Pneumonia images count in training set: {COUNT_PNEUMONIA}")
```

결과를 보니 폐렴 이미지가 3배 더 많이 있어 imbalance하다. 이번에 사용하는 CNN 모델은 balance가 좋을수록 training을 잘 하므로 차후에 조정하자. test 데이터셋은 평가하기 위해 있고 학습과 관련 없으므로 문제 없다.

## 2. tf.data로 인스턴스 만들고 전처리하기
mini batch로 학습시킬 예정이다. [mini-batch는 왜 사용하는가?](https://lms.aiffel.io/steps2/3407)에서 mini batch에 대해 알 수 있다.

```python
train_list_ds = tf.data.Dataset.from_tensor_slices(train_filenames)
val_list_ds = tf.data.Dataset.from_tensor_slices(val_filenames)
```

```python
TRAIN_IMG_COUNT = tf.data.experimental.cardinality(train_list_ds).numpy()
print(f"Training images count: {TRAIN_IMG_COUNT}")

VAL_IMG_COUNT = tf.data.experimental.cardinality(val_list_ds).numpy()
print(f"Validating images count: {VAL_IMG_COUNT}")
```

이미지 경로에 NORMAL, PNEUMONIA로 나눠져 있기 때문에 이를 이용해서 라벨 데이터 만들기

```python
# 파일 경로의 끝에서 두번째 부분인 NORMAL, PNEUMONIA를 보면 양성, 음성을 확인할 수 있다.
def get_label(file_path):
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2] == "PNEUMONIA"   # 폐렴이면 양성(True), 노말이면 음성(False)
```

이미지 데이터의 사이즈가 현실적으로 제각각일 가능성이 높아 이미지 형식 변환 및 라벨링하는 함수 만들자.

```python
# 이미지 형식 변환하는 함수
def decode_img(img):
    img = tf.image.decode_jpeg(img, channels=3) # 이미지를 uint8 tensor로 수정
    img = tf.image.convert_image_dtype(img, tf.float32) # float32 타입으로 수정
    img = tf.image.resize(img, IMAGE_SIZE) # 이미지 사이즈를 IMAGE_SIZE로 수정
    return img

# 이미지 파일의 경로를 입력하면 이미지와 라벨을 읽어오는 함수
def process_path(file_path):
    label = get_label(file_path) # 라벨 검출
    img = tf.io.read_file(file_path) # 이미지 읽기
    img = decode_img(img) # 이미지를 알맞은 형식으로 수정
    return img, label
```

```python
train_ds = train_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
val_ds = val_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
```

num_parallel_calls 인자는 병렬 데이터셋을 불러오고, 파일 여는데 시간을 단축하는 장점이 있다.

적용한거 확인해보기

```python
for image, label in train_ds.take(1): # train_ds에서 take()써서 데이터 하나 가져오기
    print("Image shape: ", image.numpy().shape)
    print("Label: ", label.numpy())
```

test 데이터셋도 적용하기

```python
test_list_ds = tf.data.Dataset.list_files(TEST_PATH) # TEST_PATH에 있는 데이터 리스트 불러오기
TEST_IMAGE_COUNT = tf.data.experimental.cardinality(test_list_ds).numpy() # 개수 카운트
test_ds = test_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
test_ds = test_ds.batch(BATCH_SIZE)

print(TEST_IMAGE_COUNT)
```

## 3. 훈련 준비하기

`tf.data` 파이프라인을 사용해서 데이터 학습시키기

```python
def prepare_for_training(ds, shuffle_buffer_size=1000):
    ds = ds.shuffle(buffer_size=shuffle_buffer_size) 
    # buffer_size에 1000이 들어가는데 이는 해당 버퍼에서 무작위로 다음 요소를 선택한다.
    ds = ds.repeat() # 반복하는 만큼 데이터를 맞춰준다.
    ds = ds.batch(BATCH_SIZE) # BATCH_SIZE에 맞춰 데이터를 나눈다. 위에서 선언한 값은 16
    ds = ds.prefetch(buffer_size=AUTOTUNE) # 학습 데이터를 나눠서 읽어와 GPU로 학습, CPU로 준비처리한다.
    return ds

train_ds = prepare_for_training(train_ds)
val_ds = prepare_for_training(val_ds)
```

## 4. 시각화

```python
# 이미지 배치를 입력하면 여러장의 이미지를 보여준다.
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10,10))
    for n in range(BATCH_SIZE):
        ax = plt.subplot(4,math.ceil(BATCH_SIZE/4),n+1)
        plt.imshow(image_batch[n])
        if label_batch[n]:
            plt.title("PNEUMONIA")
        else:
            plt.title("NORMAL")
        plt.axis("off")


image_batch, label_batch = next(iter(train_ds))
show_batch(image_batch.numpy(), label_batch.numpy())
```

train의 첫 번째 batch를 추출한다. 이는 BATCH_SIZE에 영향을 받는점 참고하자. 추출된 배치를 image와 label 데이터 셋으로 나눈다.

## 5. CNN 모델링
[CNN, Convolutional Neural Network 요약](http://taewan.kim/post/cnn/)

이미지 처리에 특화된 CNN 딥러닝 모델을 통해 모델링을 하고 결과를 만들어보자.

Convolution block 먼저 만들어보자.

```python
def conv_block(filters):
    block = tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ])
    
    return block
```

Dense block도 선언하자.

```python
def dense_block(units, dropout_rate):
    block = tf.keras.Sequential([
        tf.keras.layers.Dense(units, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(dropout_rate)
    ])
    
    return block
```

전형적인 CNN 모델과는 다르게 BatchNormalization과 Dropout을 같이 사용하고 있다. 이 방법은 많은 얘기가 있는데 논문을 확인해보자.   
- [Understanding the Disharmony between Dropout and Batch Normalization by Variance Shift](https://openaccess.thecvf.com/content_CVPR_2019/papers/Li_Understanding_the_Disharmony_Between_Dropout_and_Batch_Normalization_by_Variance_CVPR_2019_paper.pdf)   

위 논문은 variance shift를 억제하는 batch normalization과 이를 유발하는 dropout을 동시에 사용하는 것이 어울리지 않다고 한다. 하지만 동시에 같이 사용하는 것이 좋다고 하는 논문도 있다.   
-[Rethinking the Usage of Batch Normalization and Dropout in the Training of Deep Neural Networks](https://arxiv.org/pdf/1905.05928.pdf)   

이번 사례에서는 어떻게 적용될지 확인해보자.

```python
def build_model():
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),
        
        tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
        tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
        tf.keras.layers.MaxPool2D(),
        
        conv_block(32),
        conv_block(64),
        
        conv_block(128),
        tf.keras.layers.Dropout(0.2),
        
        conv_block(256),
        tf.keras.layers.Dropout(0.2),
        
        tf.keras.layers.Flatten(),
        dense_block(512, 0.7),
        dense_block(128, 0.5),
        dense_block(64, 0.3),
        
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    
    return model
```

## 6. 데이터 imbalance 처리

Normal과 Pneumonia 데이터 중에서 한쪽만 많은 imabalance 현상이 있었다. 그래서 해결하는 방법으로 `Weight balancing` 이라는 테크닉이 사용된다.   
이 테크닉은 training set의 각 데이터에서 loss를 계산할 때 특정 클래스의 데이터에 더 큰 loss 값을 갖도록 가중치를 부여하는 방법이다.   
- [딥러닝에서 클래스 불균형을 다루는 방법](https://3months.tistory.com/414)   

```python
weight_for_0 = (1 / COUNT_NORMAL)*(TRAIN_IMG_COUNT)/2.0 
weight_for_1 = (1 / COUNT_PNEUMONIA)*(TRAIN_IMG_COUNT)/2.0

class_weight = {0: weight_for_0, 1: weight_for_1}

print('Weight for NORMAL: {:.2f}'.format(weight_for_0))
print('Weight for PNEUMONIA: {:.2f}'.format(weight_for_1))
```

weight_for_0 변수는 normal 이미지에 사용한 weight를 weight_for_1 변수는 pneumonia 이미지에 사용할 weight를 세팅할 때 사용한다. 이들은 전체 데이터 건수에 반비례 하도록 설정했다.

## 7. 모델 훈련

```python
with tf.device('/GPU:0'):
    model = build_model()

    METRICS = [
        'accuracy',
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
    
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=METRICS
    )
```

이미지 훈련은 GPU를 사용하기 때문에 이를 선택했다. 모델은 위에서 먼저 선언한 build_model에 선언했다. 이미지의 라벨이 두 개밖에 없기 때문에 binary_cross entropy loss를 사용하고 optimizer는 adam을 사용했다. 성과는 metrics의 precision과 recall을 사용한다.

```python
with tf.device('/GPU:0'):
    history = model.fit(
        train_ds,
        steps_per_epoch=TRAIN_IMG_COUNT // BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=val_ds,
        validation_steps=VAL_IMG_COUNT // BATCH_SIZE,
        class_weight=class_weight,
    )
```

## 8. 결과 확인

그래프로 에포크마다 precision, recall, accuracy, loss의 변화를 확인해보자.

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 3))
ax = ax.ravel()

for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
    ax[i].plot(history.history[met])
    ax[i].plot(history.history['val_' + met])
    ax[i].set_title('Model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['train', 'val'])
```

테스트 데이터로 모델 평가를 진행해보자.

```python
loss, accuracy, precision, recall = model.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}')
```

폐렴 진단에 있어서 80%는 맞출 수 있게 되었다. 의료 보조 용도로는 어느 정도 쓸만한 모델이라 할 수 있다.   
성능 개선할 수 있는 방법은 없을까? 의료 데이터의 경우 Data augmentation 방법을 많이 시도한다. 근데 이는 각 데이터마다 최적화된 방법을 찾기 어렵고 제약사항이 많다. 그래서 의료 분야는 이미지 회전, 가우시안 노이즈 추가 등 다양한 방법을 사용한다. 특히 장기의 크기는 사람마다 크게 바뀌지 않는데 이런 특성을 이용한 방법이 노이즈 추가 방법이다. 그 외에 GAN을 이용한 Data augmentation을 시도하는 연구도 많으니 참고하자.
